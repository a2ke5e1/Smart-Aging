```python
import pandas as pd
import numpy as np
```

```python
# load dataset
df = pd.read_csv('updated_tulum.csv')
df = df.replace(np.nan, 0)
df
```

```python
# loading only sensor data
feature = df.iloc[:, 2:7]
feature
```

```python
# annotations 
labels = df.loc[:, 'R1_Annotation']
labels
```

```python
from sklearn import svm 
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
train, test, train_labels, test_labels = train_test_split(feature,labels,test_size=0.1,random_state=4)
```

```python
gnb = tree.DecisionTreeClassifier()
model = gnb.fit(train, train_labels)
print("model",model)
```

```python
pred = model.predict(test)
pred
```

```python
accuracy_score = accuracy_score(test_labels, pred)

print("Accuracy: ", accuracy_score)
```

```python
# classification report
from sklearn.metrics import classification_report

print(classification_report(test_labels, pred))
```
