```python
import pandas as pd
import numpy as np
```

```python
# load dataset
df = pd.read_csv('updated_tulum.csv')
df = df.replace(np.nan, 0)
df
```

```python
# loading only sensor data
feature = df.iloc[:, 2:22]
feature
```

```python
# annotations 
labels = df.loc[:, 'R1_Annotation':'R2_Annotation']
labels
```

```python
from sklearn import svm 
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
train, test, train_labels, test_labels = train_test_split(feature,labels,test_size=0.1,random_state=4)
```

```python
gnb = tree.DecisionTreeClassifier()
model = gnb.fit(train, train_labels)
print("model",model)
```

```python
# get first  1 row of the test data
print(test.iloc[0:1] )
```

```python
preds = gnb.predict(test)
print(preds)
```

```python
# get input shape of the model
print( "Input shape: ", test.shape )
```

```python
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType

# Specify an initial type for the model ( similar to input shape for the model )
initial_type = [('float_input', FloatTensorType([None, 20]))]


# Write the ONNX model to disk
converted_model = convert_sklearn( gnb , initial_types=initial_type )
with open( "sklearn_model_multi-people.onnx", "wb" ) as f:
    f.write( converted_model.SerializeToString())
```
